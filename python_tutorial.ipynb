1. In method overloading, Python provides the feature of creating methods that have the same name to perform or execute different functionalities in a given piece of code. It allows to overload methods and uses them to perform different tasks in simpler terms.
2. In Method overriding, Python overrides the value that shares a similar name in parent and child classes.

```python
class baseclass:
    def __init__(self, name):
        self.name = name
    def area1(self):
        pass
    def __str__(self):
        return self.name
class rectangle(baseclass):
    def __init__(self, length, breadth):
        super().__init__("rectangle")
        self.length = length
        self.breadth = breadth
    def area1(self):
        return self.length * self.breadth
class triangle(baseclass):
    def __init__(self, height, base):
        super().__init__("triangle")
        self.height = height
        self.base = base
    def area1(self):
        return (self.base * self.height) / 2
a = rectangle(90, 80)
b = triangle(77, 64)
print("The shape is: ", b)
print("The area of shape is", b.area1())
print("The shape is:", a)
print("The area of shape is", a.area1())
```

## 6.3 GENERATOR AND YIELD VS. RETURN

### WHAT IS PYTHON YIELD
The yield keyword in python works like a return with the only difference is that instead of returning a value, it gives back a generator object to the caller.

When a function is called and the thread of execution finds a yield keyword in the function, the function execution stops at that line itself and it returns a generator object back to the caller.

Python yield returns a generator object. Generators are special functions that have to be iterated to get the values.

The yield keyword converts the expression given into a generator function that gives back a generator object. To get the values of the object, it has to be iterated to read the values given to the yield.

### Example: Yield Method
Here is a simple example of yield. The function testyield() has a yield keyword with the string â€œWelcome to my Python Tutorialâ€. When the function is called, the output is printed and it gives a generator object instead of the actual value.

```python
def testyield():
  yield "Welcome to my Python Tutorial"
output = testyield()
print(output)
```

The output of the above code is a generator object, which has the value we have given to yield.

But we are not getting the message we have to given to yield in output!

To print the message given to yield will have to iterate the generator object as shown in the example below:

```python
def testyield():
  yield "Welcome to my Python Tutorial"

output = testyield()
for i in output:
    print(i)
```

### What are Generators in Python?

Generators are functions that return an iterable generator object. The values from the generator object are fetched one at a time instead of the full list together and hence to get the actual values you can use a for-loop, using next() or list() method.

### Using Generator function

You can create generators using generator function and using generator expression.

A generator function is like a normal function, instead of having a return value it will have a yield keyword.

To create a generator function you will have to add a yield keyword. 

```python
#Example to create a generator function
def generator():
    yield "C"
    yield "S"
    yield "U"
    yield "L"
    yield "B"

test = generator()
for i in test:
    print(i)
```

### Difference between Normal function v/s Generator function.
Let us understand how a generator function is different from a normal function.

There are 2 functions normal_test() and generator_test().

Both the functions are suppose to return back the string "CSULB RECORDSâ€. The normal_test() is using return and generator_test() is using yield.

The output of the following code will show that when you call the normal function normal_test() it returns CSUL RECORDS string. For a generator function with yield keyword it returns <generator object generator_test at 0x7fb376bf8430> and not the string.

This is the main difference between a generator function and a normal function. Now to get the value from the generator object we need to either use the object inside for loop or use next() method or make use of list().

One more difference to add to normal function v/s generator function is that when you call a normal function the execution will start and stop when it gets to return and the value is returned to the caller. So when the execution starts you cannot stop the normal function in between and it will only stop when it comes across return keyword.

But in case of generator function once the execution starts when it gets the first yield it stops the execution and gives back the generator object. You can use the generator object to get the values and also, pause and resume back as per your requirement.

```python
# Normal function
def normal_test():
    return "CSULB RECORDS"
	
#Generator function
def generator_test():
	yield "CSULB RECORDS"
print(normal_test()) #call to normal function
print(generator_test()) # call to generator function
print(next(generator_test()))  # will output CSULB RECORDS
```

### How to read the values from the generator?
You can read the values from a generator object using a list(), for-loop and using next() method.

```python
#Using : list()
#A list is an iterable object that has its elements inside brackets.Using list() on a generator object will give all the values the generator holds.

def even_numbers(n):
    for x in range(n):
       if (x%2==0): 
           yield x       
num = even_numbers(10)
print(list(num))
```

```python
#Using : for-in
#In the example, there is a function defined even_numbers() that will give you all even numbers for the n defined. The call to the function even_numbers() will return a generator object, that is used inside for-loop.

def even_numbers(n):
    for x in range(n):
       if (x%2==0): 
           yield x       
num = even_numbers(10)
for i in num:
    print(i)
```

```python
#Using next()
#The next() method will give you the next item in the list, array, or object. Once the list is empty, and if next() is called, it will give back an error with stopIteration signal. This error, from next() indicates that there are no more items in the list.

def even_numbers(n):
    for x in range(n):
       if (x%2==0): 
           yield x       
num = even_numbers(10)
print(next(num))
print(next(num))
print(next(num))
print(next(num))
print(next(num))
```

### Generators are one-time Use
Incase of generators they are available for use only once. If you try to use them again, it will be empty. The following example shows that generators are one time use:

```python
def even_numbers(n):
    for x in range(n):
       if (x%2==0): 
           yield x       
num = even_numbers(10)
for i in num:
    print(i)

print("\n")
print("Calling the generator again: ", list(num))
```

```python
#The following example shows how to use generators and yield in Python. The example will generate the Fibonacci series
def getFibonnaciSeries(num):
    c1, c2 = 0, 1
    count = 0
    while count < num:
        yield c1
        c3 = c1 + c2
        c1 = c2
        c2 = c3
        count += 1
fin = getFibonnaciSeries(7)
print(fin)
for i in fin:
    print(i)
```

### When to use Yield Instead of Return in Python
Python3 Yield keyword returns a generator to the caller and the execution of the code starts only when the generator is iterated.

A return in a function is the end of the function execution, and a single value is given back to the caller.

Here, is the situation when you should use Yield instead of Return:

Use yield instead of return when the data size is large
Yield is the best choice when you need your execution to be faster on large data sets.
Use yield when you want to return a big set of values to the calling function
Yield is an efficient way of producing data that is big or infinite.

## 6.7 Break, Continue and Pass Statements

The concept of loops is available in almost all programming languages. Python loops help to iterate over a list, tuple, string, dictionary, and a set. There are two types of loop supported in Python â€œforâ€ and â€œwhileâ€. The block of code is executed multiple times inside the loop until the condition fails.

The loop control statements break the flow of execution and terminate/skip the iteration as per our need. Python break and continue are used inside the loop to change the flow of the loop from its standard procedure.

A for-loop or while-loop is meant to iterate until the condition given fails. When you use a break or continue statement, the flow of the loop is changed from its normal way.

### Python break statement
The break statement takes care of terminating the loop in which it is used. If the break statement is used inside nested loops, the current loop is terminated, and the flow will continue with the code followed that comes after the loop.

### Break statement execution flow
When the for-loop starts executing, it will check the if-condition. If true, the break statement is executed, and the forâ€“loop will get terminated. If the condition is false, the code inside for-loop will be executed.

When the while loop executes, it will check the if-condition; if it is true, the break statement is executed, and the while â€“loop will exit. If if the condition is false, the code inside while-loop will get executed.



### Example: Break statement inside for-loop
The list my_list = [â€˜Ryanâ€™, â€˜Tanyaâ€™, â€˜Grahamâ€™, â€˜Davidâ€™, â€˜Riyaâ€™, â€˜Davidâ€™] is looped using for-loop.We are interested in searching for the name â€˜David â€˜ from the list my_list.

Inside the for-loop, the if-condition compares each item from the list with the name â€˜Davidâ€™. If the condition becomes true, it will execute the break statement, and the loop will get terminated.

The working example using break statement is as shown below:

```python
my_list = ['Ryan', 'Tanya', 'Graham', 'David', 'Riya', 'David'] 

for i in range(len(my_list)):
    print(my_list[i])
    if my_list[i] == 'David':
        print('Found the name David')
        break
        print('After break statement')

print('Loop is Terminated')
```

```python
#Example Break statement inside while loop
my_list = ['Ryan', 'Tanya', 'Graham', 'David', 'Riya', 'David'] 
i = 0

while True:
    print(my_list[i])
    if (my_list[i] == 'David'):
        print('Found the name David')
        break
        print('After break statement')
    i += 1

print('After while-loop exit')
```

So because of the break statement, the second for-loop will never iterate for 2 and 3.

```python
for i in range(4):
    for j in range(4):          
        if j==2:    
            break
        print("The number is ",i,j);  
```

When the while loop executes, it will check the if-condition, if it is true, the continue statement is executed. The control will go back to the start of while â€“loop for the next iteration. If if the condition is false, the code inside while-loop will get executed.

## 9.1 Modules:
A module in Python refers to a file that includes Python statements and definitions. The module name is the same as the file name with a .py extension added to it. By importing modules into other Python programs, you can recycle code and create more sophisticated programs.

### Creating a module:
To create a module, create a new python file and add your desired functions/statements. For example, we want to create a module that calculates the grade point average (GPA) of a CSULB student. Start by creating a new Python file called csulb_gpa.py and define a function called calculate_gpa() that takes a list of grades as an input and returns the GPA:


```python
# Contents of the file - csulb_gpa.py

def calculate_gpa(grades):
    total_units = 0
    total_grade_points = 0

    for grade, units in grades:
        if grade == 'A':
            grade_point = 4.0
        elif grade == 'A-':
            grade_point = 3.7
        elif grade == 'B+':
            grade_point = 3.3
        elif grade == 'B':
            grade_point = 3.0
        elif grade == 'B-':
            grade_point = 2.7
        elif grade == 'C+':
            grade_point = 2.3
        elif grade == 'C':
            grade_point = 2.0
        elif grade == 'C-':
            grade_point = 1.7
        elif grade == 'D+':
            grade_point = 1.3
        elif grade == 'D':
            grade_point = 1.0
        else:
            grade_point = 0.0

        total_units += units
        total_grade_points += grade_point * units

    gpa = total_grade_points / total_units
    return gpa
```

### Importing a module:
Once the csulb_gpa.py module has been created, it can be imported into another Python program using the import statement. To import a module, use the import statement in a new python file. For example, to calculate the GPA of a CSULB student, we will create a new python file student_gpa.py. 


```python
# Contents of file - student_gpa.py
import csulb_gpa

grades = [('A', 3), ('B', 4), ('C+', 2), ('A-', 3)]
gpa = csulb_gpa.calculate_gpa(grades)
print('The GPA is:', gpa)
```

## 12.3 Using Numpy

### Creating nd-arrays
An nd-array is a multidimensional array that can be used to store and manipulate large amounts of data. In Numpy, nd-arrays can be created using the numpy.array() function. Here is an example of creating an nd-array:

```python
# creating a 2x3 array
first_array = np.array([[1, 2, 3], [4, 5, 6]])
print(first_array)
```

### Vectors and matrices
Numpy provides support for creating vectors and matrices using the numpy.array() function. Here is an example:

```python
# create a vector
first_vector = np.array([1, 2, 3])
print('vector:\n')
print(first_vector)

# create a matrix
third_matrix = np.array([[1, 2, 3], [4, 5, 6]])
print('\nmatrix:\n')
print(third_matrix)
```

### Boolean indexing
Boolean indexing is a powerful technique that allows us to filter nd-arrays based on a condition. Here is an example:

```python
# create a 1D array
second_array = np.array([1, 2, 3, 4, 5])
print('Array before filter\n')
print(second_array)
# create a boolean array based on the condition
bool_array = second_array > 3

# filter the array based on the boolean array
filtered_array = second_array[bool_array]
print('\nArray after filter\n')
print(filtered_array)
```

### Reshaping data
Reshaping data is an important operation in Numpy that allows us to change the shape of an nd-array without changing its data. Here is an example:



```python
# create a 1D array with 12 elements
third_array = np.arange(12)

# reshape the array into a 3x4 matrix
first_matrix = third_array.reshape((3, 4))
print('Matrix 3X4\n')
print(first_matrix)

second_matrix = third_array.reshape(4,3)
print('\nMatrix 4X3\n')
print(second_matrix)
```

## 12.4 Using Matplotlib

### Plotting your first graph
Matplotlib is a data visualization library that can be used to create various types of plots. Let's start by plotting a simple line graph using Matplotlib.

```python
import matplotlib.pyplot as plt

# create some data
x = np.array([0, 1, 2, 3, 4])
y = np.array([1, 3, 2, 4, 3])

# plot the data
plt.plot(x, y)

# add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Graph')

# show the plot
plt.show()
```

### Multiplots and sub-plots
Matplotlib also provides the capability to create multiple plots in a single figure using sub-plots.

```python
# create some data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# create subplots
fig, (ax1, ax2) = plt.subplots(1, 2)

# plot data on each subplot
ax1.plot(x, y1)
ax2.plot(x, y2)

# add labels and title to each subplot
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Y-axis')
ax1.set_title('Sin Graph')
ax2.set_xlabel('X-axis')
ax2.set_ylabel('Y-axis')
ax2.set_title('Cos Graph')

# adjust layout and show the plot
plt.tight_layout()
plt.show()
```

### Various types of plots
Matplotlib provides a variety of plot types, including bar plots, scatter plots, and histograms. Let's see some examples of these plots.

#### Line plot
A line plot is used to visualize data as a series of points connected by a straight line. It is useful for showing trends over time or other continuous data. Here is an example of a line plot:

```python
# Generate some data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a line plot
plt.plot(x, y)

# Add labels and title
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Sinusoidal Wave')
plt.show()
```

#### Scatter Plot
A scatter plot is used to visualize the relationship between two variables. It is useful for identifying patterns and outliers in the data. Here is an example of a scatter plot:

```python
# Generate some data
x = np.random.randn(100)
y = x + np.random.randn(100) * 0.5

# Create a scatter plot
plt.scatter(x, y)

# Add labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot')
plt.show()
```

#### Bar Plot
A bar plot is used to compare values across different categories. It is useful for showing the relative size or frequency of different groups. Here is an example of a bar plot:

```python
# Generate some data
labels = ['Math', 'Science', 'English', 'History']
values = [85, 72, 92, 80]

# Create a bar plot
plt.bar(labels, values)

# Add labels and title
plt.xlabel('Subject')
plt.ylabel('Grade')
plt.title('CSULB Student Grades')
plt.show()
```

#### Histogram
A histogram is used to visualize the distribution of a single variable. It is useful for showing the range and frequency of values in the data. Here is an example of a histogram:

```python
# Generate some data
x = np.random.randn(1000)

# Create a histogram
plt.hist(x, bins=30)

# Add labels and title
plt.xlabel('X')
plt.ylabel('Frequency')
plt.title('Histogram of Random Numbers')
plt.show()
```

#### Pie Chart
A pie chart is used to show the proportion of different categories in the data. It is useful for displaying percentages and relative sizes of different groups. Here is an example of a pie chart:

```python
# Generate some data
labels = ['Math', 'Science', 'English', 'History']
sizes = [20, 30, 25, 25]

# Create a pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)

# Add title
plt.title('CSULB Student Majors')

# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')

# Show plot
plt.show()
```

#### Box plot
Box plot, also known as box and whisker plot, is a useful tool for visualizing the distribution of data. It shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution. The plot consists of boxes that represent the first, second (median), and third quartiles. The whiskers are the lines that extend from the box to the minimum and maximum values, excluding any outliers.

Here's an example of a box plot:

```python
# Generate some random data
np.random.seed(42)
data = np.random.normal(size=(100, 5))

# Create a box plot
plt.boxplot(data, labels=['A', 'B', 'C', 'D', 'E'])

# Add title and axis labels
plt.title('Box plot of randomly generated data')
plt.xlabel('Category')
plt.ylabel('Value')

# Show the plot
plt.show()
```

#### Heat map
Heat maps are used to visualize data as a color-coded matrix. They are useful when you need to quickly understand the relationship between two variables. The color scheme used can vary, but typically red is used to represent high values and blue to represent low values.

Here's an example of a heat map:

```python
# Generate some random data
np.random.seed(42)
data = np.random.rand(10, 10)

# Create a heat map
plt.imshow(data, cmap='coolwarm')

# Add color bar and axis labels
plt.colorbar()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show the plot
plt.show()
```

#### 3D plot
Matplotlib also allows us to create 3D plots to visualize data in three dimensions. 3D plots can be created using the mplot3d toolkit, which is included in the mpl_toolkits package.

Here's an example of a 3D plot:

```python
from mpl_toolkits.mplot3d import Axes3D

# Generate some random data
np.random.seed(42)
x = np.random.normal(size=500)
y = np.random.normal(size=500)
z = x**2 + y**2

# Create a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, c=z, cmap='coolwarm')

# Add title and axis labels
ax.set_title('3D scatter plot of random data')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

# Show the plot
plt.show()
```
